# Dependencies loading

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

import warnings
warnings.filterwarnings('ignore')  # Suppressing warning
```

# Data loading

```python
base_url = "https://raw.githubusercontent.com/Niyetali/FIFA-Rating-Prediction/main/output"

# importing dataset
models_summary = pd.read_csv(f"{base_url}/final.csv", sep=',')
models_summary.fillna('-', inplace=True)
```

# Final Model Comparison

```python
plt.figure(figsize=(8, 5))
plt.bar(models_summary["Model"], models_summary["MAE"], color=['blue', 'orange', 'green', 'purple'])
plt.title("MAE Comparison Across Models")
plt.ylabel("MAE")
plt.xlabel("Model")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.bar(models_summary["Model"], models_summary["MSE"], color=['blue', 'orange', 'green', 'purple'])
plt.title("MSE Comparison Across Models")
plt.ylabel("MSE")
plt.xlabel("Model")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.bar(models_summary["Model"], models_summary["RMSE"], color=['blue', 'orange', 'green', 'purple'])
plt.title("RMSE Comparison Across Models")
plt.ylabel("RMSE")
plt.xlabel("Model")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.bar(models_summary["Model"], models_summary["MAPE"], color=['blue', 'orange', 'green', 'purple'])
plt.title("MAPE Comparison Across Models")
plt.ylabel("MAPE")
plt.xlabel("Model")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

plt.figure(figsize=(8, 5))
plt.bar(models_summary["Model"], models_summary["SMAPE"], color=['blue', 'orange', 'green', 'purple'])
plt.title("SMAPE Comparison Across Models")
plt.ylabel("SMAPE")
plt.xlabel("Model")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.bar(models_summary["Model"], models_summary["R2"], color=['blue', 'orange', 'green', 'purple'])
plt.title("R Score Comparison Across Models")
plt.ylabel("R Score")
plt.xlabel("Model")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.bar(models_summary["Model"], models_summary["Adj R2"], color=['blue', 'orange', 'green', 'purple'])
plt.title("Adjusted R Score Comparison Across Models")
plt.ylabel("Adjusted R Score")
plt.xlabel("Model")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Rounding MSE, MAE, RMSE to two decimal places
models_summary['MSE'] = models_summary['MSE'].round(2)
models_summary['MAE'] = models_summary['MAE'].round(2)
models_summary['RMSE'] = models_summary['RMSE'].round(2)

# Converting MAPE, SMAPE, R2, and Adj R2 to percentages with two decimal places
models_summary['MAPE'] = (models_summary['MAPE'] * 100).round(2).astype(str) + '%'
models_summary['SMAPE'] = (models_summary['SMAPE'] * 100).round(2).astype(str) + '%'
models_summary['R2'] = (models_summary['R2'] * 100).round(2).astype(str) + '%'
models_summary['Adj R2'] = (models_summary['Adj R2'] * 100).round(2).astype(str) + '%'

models_summary
```

# Overview

The models show strong predictive abilities, but they differ in performance metrics. Here's the breakdown:

###  **OLS (Ordinary Least Squares)**  - Baseline
As the baseline model, OLS delivers a solid R of **0.9297** but falls short with higher errors:  
- **MSE**: 4.4915  
- **RMSE**: 2.1193  
- **MAE**: 1.6565  

While decent, it struggles to minimize errors effectively.  

---

###  **AdaBoost**  
After tuning, AdaBoost takes a big leap in accuracy!   
- **R**: 0.9872  
- **MSE**: 0.8147  
- **MAE**: 0.4871  
- **MAPE**: 0.823%  

Its performance proves its highly reliable for precise predictions.   

---

###  **XGBoost**  
The star of the show ! XGBoost boasts:  
- **R**: 0.9876 (highest)  
- **MSE**: 0.7905 (lowest)  
- **RMSE**: 0.8891  
- **MAE**: 0.5525  

Although its MAE is slightly higher than AdaBoost, it shines with an excellent balance between precision and error minimization.  

---

###  **MLP (Multi-Layer Perceptron)**  
MLP delivers decent results:  
- **R**: 0.9733  
- **MSE**: 1.7014  
- **RMSE**: 1.3043  
- **MAE**: 0.9014  

It performs well but falls behind the ensemble-based models in accuracy and error reduction.  

---

###  **Takeaway**  
- **XGBoost** leads the pack, with **AdaBoost** close behind.  
- **OLS** and **MLP** serve as respectable but less precise options.

#  Additional Insights

While feature selection after hyperparameter tuning didnt improve the results, the models themselves did a great job capturing the most important relationships through their unique hyperparameters:  

- **XGBoost**: Key hyperparameters like **learning rate**, **max depth**, and **subsample** helped optimize performance by fine-tuning how the model learns and generalizes.  
- **AdaBoost**: Parameters such as the **number of estimators** and **learning rate** ensured that the ensemble model effectively combined weak learners to achieve high accuracy.  
- **MLP (Multi-Layer Perceptron)**: Network-specific hyperparameters like the **number of hidden layers**, **neurons per layer**, and **activation functions** enabled it to model complex, non-linear relationships in the data.  

Earlier steps like **feature engineering** and **dataset refinement** ensured that all models worked with the most relevant and impactful data, further enhancing their performance.
